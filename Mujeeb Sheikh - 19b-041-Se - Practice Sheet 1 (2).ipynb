```python
# Question No 7: Three resiters of 10, 20 and 30 ohms are connected in series and in parallel if 10 V is given to them
# find current and voltages across them.

R1=10
R2=20
R3=30
V=10 
Rt=R1+R2+R3
I=V/Rt
V1=I*R1
V2=I*R2
V3=I*R3
print("Total Current across series circuit: ",I)
print("Voltage across R1:",V1 , 'V')
print("Voltage across R2:",V2 , 'V')
print("Voltage across R3:",V3 , 'V')
print()

#Resistors in Parallel
TotResistance=1/(1/R1 +1/R2 +1/R3)
TotCurrent=V/TotResistance
I1=V/R1
I2=V/R2
I3=V/R3

print("Total current accross parallel circuit: ",TotCurrent)
print("Current across R1:",I1 ,'Amp' )
print("Current across R2:",I2 ,'Amp')
print("Current across R3:",I3 ,'Amp' )


print("Time taken : 10 mins")
```

```python
# Question 8 : Find the point of intercept if line 1 coordinates are (1,5), (7, 9) and line 2 coordinates are (2,4) (6,9).

# For Line 1
x1 = 1
x2 = 7
y1 = 7 
y2 = 9

a1 = (y2-y1)/(x2-x1)
print ("Gradient of line 1 : " , a1)

C2 = y1-a1*x1
print ("y intercept of line 1 : " , C2)
print ("Equation of Line 1 : 0.6x - y + 4.3 = 0 ")

# For line 2

x3 = 6
x4 = 2
Y3 = 9
y4 = 4

a2 = (y4-Y3)/(x4-x3)
print ("\nGradient of line 2 : " , a2)

C3 = Y3-a2*x3
print ("y intercept of line 1 : " , C3)
print ("Equation of Line 2 : 1.25x - y + 1.5 = 0 ")

xintercept = (C3-C2)/(a2-a1)
yintercept = (C2*a2-C3*a1)/(a2-a1)

print ("\n\n at the point of intersection X Cordinate is : " , xintercept)
print ("at the point of intersection Y intercept is : " , yintercept)

print("Time Taken 30min")




```
